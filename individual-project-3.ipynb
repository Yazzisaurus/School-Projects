```python
#import required libraries
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn import preprocessing
```

```python
#read CSV
df = pd.read_csv("/kaggle/input/credit-fraud/credit_fraud.csv")
df.head()
```

```python
#Define the index
columns = ['id','over_draft','credit_usage','credit_history','purpose','current_balance','Average_Credit_Balance','employment','location','personal_status','other_parties', 'residence_since','property_magnitude','cc_age', 'other_payment_plans', 'housing','existing_credits','job', 'num_dependents', 'own_telephone', 'foreign_worker','class']
df =df.iloc[301:,:]
df=df.reset_index(drop=True)
df.columns=columns
print('Shape: ',df.shape)
df.head()
```

```python
#checking the data type
df.info()
```

```python
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
df['over_draft'] = le.fit_transform(df['over_draft'])
df['credit_history'] = le.fit_transform(df['credit_history'])
df['purpose'] = le.fit_transform(df['purpose'])
df['Average_Credit_Balance'] = le.fit_transform(df['Average_Credit_Balance'])
df['employment'] = le.fit_transform(df['employment'])
df['personal_status'] = le.fit_transform(df['personal_status'])
df['other_parties'] = le.fit_transform(df['other_parties'])
df['property_magnitude'] = le.fit_transform(df['property_magnitude'])
df['other_payment_plans'] = le.fit_transform(df['other_payment_plans'])
df['housing'] = le.fit_transform(df['housing'])
df['job'] = le.fit_transform(df['job'])
df['own_telephone'] = le.fit_transform(df['own_telephone'])
df['foreign_worker'] = le.fit_transform(df['foreign_worker'])
df['class'] = le.fit_transform(df['class'])
```

```python
df['id'] = [int(row) for row in df['id']]
df['credit_usage'] = [int(row) for row in df['credit_usage']]
df['current_balance'] = [int(row) for row in df['current_balance']]
df['location'] = [int(row) for row in df['location']]
df['residence_since'] = [int(row) for row in df['residence_since']]
df['cc_age'] = [int(row) for row in df['cc_age']]
df['existing_credits'] = [int(row) for row in df['existing_credits']]
df['num_dependents'] = [int(row) for row in df['num_dependents']]
```

```python
#Defining the Gaussian curve
print('Shape: ',df.shape)
df.head()
```

```python
# splitting the data
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1].values, df.iloc[:,-1:].values,
                                                    test_size=0.2, random_state=0)
```

```python
#Modeling 
#Use sklearn package to model data
clf = GaussianNB().fit(X_train, y_train)

#The coefficients
y_pred = clf.predict(X_test)
print('Accuracy score: ',accuracy_score(y_test, y_pred))
```
